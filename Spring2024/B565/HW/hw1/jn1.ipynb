```python
def calculate_euclidean_dist(arr1, arr2):
    return np.sqrt(np.sum((arr1 - arr2) ** 2))
```

```python
d[10:, :].shape
```

```python
def find_min_max_dist(d):
    if d.shape[1] < 2:
        return 0, 0
    d_max = sys.float_info.min
    d_min = sys.float_info.max
    for i, point_i in enumerate(d):
        for j, point_j in enumerate(d[i+1:, : ]):
            # if i == j:
            #     continue
            # else:
            dist = calculate_euclidean_dist(point_i, point_j)
            d_max = (
                dist
                if dist > d_max
                else d_max
            )
            d_min = (
                dist
                if dist < d_min
                else d_min
            )

    return d_min, d_max
```

```python
# n = [150, 1500, 15000, 150000]
epoch = 10
n = [150, 1500]
m = list(range(1, 101))

# f_m_execution_times = list()
# fm_creation_execution_times = list()

f_m_execution_times_multiple = list()
fm_creation_execution_times_multiple = list()
all_fms = list()

for k in range(epoch):
    f_m_execution_times = list()
    fm_creation_execution_times = list()

    for i in n:
        _start = time.time()
        f_m = np.zeros(len(m))
        _stop = time.time()
        fm_creation_execution_time = _stop - _start

        f_m_counter = 0
        f_m_execution_times_n = list()
        for j in m:
            loop_start = time.time()

            _start = time.time()
            d = np.random.uniform(0, 1, (i, j))
            _stop = time.time()
            d_creation_execution_time = _stop - _start

            _start = time.time()
            d_min, d_max = find_min_max_dist(d)
            _stop = time.time()
            d_min_max_execution_time = _stop - _start

            try:
                f_m[f_m_counter] = np.log10(
                    np.array(((d_max - d_min) / d_min), dtype=np.float64)
                )
            except ZeroDivisionError:
                f_m[f_m_counter] = 0
            f_m_counter += 1

            loop_end = time.time()
            fm_execution_time = loop_end - loop_start
            f_m_execution_times_n.append(fm_execution_time)

            with open("jn1_test_out.txt", "a") as f:
                f.write(
                    f"epoch={k}, n={i}, f_m_time={fm_execution_time} m={j}, d_time={d_creation_execution_time} d_min={d_min}, d_max={d_max}, d_min_max_time={d_min_max_execution_time} f(m)={f_m[f_m_counter - 1]}\n"
                )

            print(
                f"epoch={k}, n={i}, f_m_time={fm_execution_time} m={j}, d_time={d_creation_execution_time} d_min={d_min}, d_max={d_max}, d_min_max_time={d_min_max_execution_time} f(m)={f_m[f_m_counter - 1]}"
            )
        f_m_execution_times.append(f_m_execution_times_n)
        all_fms.append(f_m)

    f_m_execution_times_multiple.append(f_m_execution_times)
```

```python
all_fms_np = np.zeros((len(n), len(m)))
print(len(all_fms))
for i in range(len(all_fms)):
    all_fms_np[i % len(n)] += all_fms[i]

all_fms_np = all_fms_np / len(n)
all_fms_np[1]
```
